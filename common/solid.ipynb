1. S — Single Responsibility Principle (Принцип єдиної відповідальності)

```python
# ❌ Погано: один клас робить усе — і логику, і збереження, і валідацію.
class User:
    def __init__(self, name, email):
        self.name = name
        self.email = email

    def validate_email(self):
        return "@" in self.email

    def save_to_db(self):
        print(f"Saving {self.name} to database")

# ✅ Добре: кожен клас має свою відповідальність
class User:
    def __init__(self, name, email):
        self.name = name
        self.email = email

class UserValidator:
    def is_valid(self, user: User):
        return "@" in user.email

class UserRepository:
    def save(self, user: User):
        print(f"Saving {user.name} to database")
```

2. O — Open/Closed Principle (Відкритість до розширення, закритість до модифікації)

```python
from abc import ABC, abstractmethod

# ✅ Ми можемо додавати нові типи знижок, не змінюючи клас Order.
class Discount(ABC):
    @abstractmethod
    def calculate(self, total):
        pass

class NoDiscount(Discount):
    def calculate(self, total):
        return total

class PercentageDiscount(Discount):
    def __init__(self, percent):
        self.percent = percent

    def calculate(self, total):
        return total * (1 - self.percent / 100)

class Order:
    def __init__(self, total, discount: Discount):
        self.total = total
        self.discount = discount

    def final_price(self):
        return self.discount.calculate(self.total)

# Новий тип знижки можна додати без зміни Order
```

3. L — Liskov Substitution Principle (Принцип підстановки Лісков)

```python
class Bird:
    def fly(self):
        return "Flying"

class Sparrow(Bird):
    pass

# ❌ Пінгвін не літає — порушує очікування базового класу
class Penguin(Bird):
    def fly(self):
        raise NotImplementedError("Penguins can't fly")

# ✅ Виправлення через зміну ієрархії
class Bird:
    pass

class FlyingBird(Bird):
    def fly(self):
        return "Flying"

class Sparrow(FlyingBird):
    pass

class Penguin(Bird):
    def swim(self):
        return "Swimming"
```

4. I — Interface Segregation Principle (Принцип розділення інтерфейсів)

```python
from abc import ABC, abstractmethod

# ❌ Один громіздкий інтерфейс для всіх роботів
class Robot(ABC):
    @abstractmethod
    def cook(self): pass
    @abstractmethod
    def clean(self): pass
    @abstractmethod
    def dance(self): pass

# ✅ Розділення інтерфейсів
class Cooking(ABC):
    @abstractmethod
    def cook(self): pass

class Cleaning(ABC):
    @abstractmethod
    def clean(self): pass

class Dancing(ABC):
    @abstractmethod
    def dance(self): pass

class KitchenRobot(Cooking, Cleaning):
    def cook(self): print("Cooking")
    def clean(self): print("Cleaning")
```

5. D — Dependency Inversion Principle (Принцип інверсії залежностей)

```python
from abc import ABC, abstractmethod

# Абстракція
class MessageSender(ABC):
    @abstractmethod
    def send(self, message): pass

# Реалізація через email
class EmailSender(MessageSender):
    def send(self, message):
        print(f"Sending email: {message}")

# Модуль верхнього рівня залежить лише від абстракції
class NotificationService:
    def __init__(self, sender: MessageSender):
        self.sender = sender

    def notify(self, message):
        self.sender.send(message)

# Ми можемо підставити будь-яку реалізацію, не змінюючи NotificationService
service = NotificationService(EmailSender())
service.notify("SOLID in action!")

# ❌ Поганий код: NotificationService напряму залежить від EmailSender
class NotificationService:
    def __init__(self):
        self.sender = EmailSender()

    def notify(self, message):
        self.sender.send(message)
```

##### Суть принципу

**Залежності повинні бути від абстракцій, а не від конкретних реалізацій.**

Ідея в тому, щоб верхній рівень логіки (**high-level**) не знав, як саме працюють дрібні деталі (**low-level**). Він має спілкуватися тільки через чітко визначені інтерфейси (абстракції).

### Переваги:

- **🔧 Гнучкість коду**: можна міняти реалізацію без змін у логіці
- **🧪 Легше тестувати**: можна підставити "фейкові" (mock) об'єкти  
- **🔗 Менше зчеплення** (low coupling): одна зміна не валить усе

> **Головна думка**: Модулі верхнього рівня не повинні залежати від модулів нижнього рівня. Обидва повинні залежати від абстракцій.
