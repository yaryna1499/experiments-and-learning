```python
import httpx
import csv
import io
```

```python
from httpx import HTTPError
```

```python
url = "https://as-metadata-services-assets.s3.amazonaws.com/sample_data/cloud_storage/eq_taq_1min.csv"
```

```python
async def _fetch_data_preview_url_content(url: str) -> str:
    try:
        async with httpx.AsyncClient(timeout=10.0) as client:
            response = await client.get(url)
            response.raise_for_status()
    except httpx.HTTPError as e:
        raise Exception(f"Failed to fetch data preview URL: {e}") from e
    return response.text
```

```python
content = await _fetch_data_preview_url_content(url)
csv_file = io.StringIO(content)
reader = csv.DictReader(csv_file)
reader.fieldnames
```
